```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('map.xlsx')
```

```python
df.head()
```

```python
# Сделаем красивый формат вида n%

for i in range(len(df['Выполнен зачет'])):
    df['Выполнен зачет'][i] = (str(int(df['Выполнен зачет'][i] * 100)) + '%')
    df['Выполнено ДЗ'][i] = (str(int(df['Выполнено ДЗ'][i] * 100)) + '%')
```

```python
df.head()
```

```python
df0=df[df['Оценка'] == 2]
df1=df[df['Оценка'] == 3]
df2=df[df['Оценка'] == 4]
df3=df[df['Оценка'] == 5]

plt.scatter(df0['Выполнено ДЗ'], df0['Выполнен зачет'], color='yellow', s=500, label='2', cmap='viridis', alpha=0.5)
plt.scatter(df1['Выполнено ДЗ'], df1['Выполнен зачет'], color='blue', s=1000, label='3', cmap='viridis', alpha=0.5)
plt.scatter(df2['Выполнено ДЗ'], df2['Выполнен зачет'], color= 'red', s=2500, label='4', cmap='viridis', alpha=0.5)
plt.scatter(df3['Выполнено ДЗ'], df3['Выполнен зачет'], color= 'green', s=3500, label='5', cmap='viridis', alpha=0.5)

plt.text(10, 12 - 0.3, "Оценка = 2", horizontalalignment='left', size='medium', color='yellow', weight='semibold')
plt.text(10, 12 - 0.6, "Оценка = 3", horizontalalignment='left', size='medium', color='blue', weight='semibold')
plt.text(10, 12 - 0.9, "Оценка = 4", horizontalalignment='left', size='medium', color='red', weight='semibold')
plt.text(10, 12 - 1.2, "Оценка = 5", horizontalalignment='left', size='medium', color='green', weight='semibold')

plt.xlabel('Выполнено ДЗ')
plt.ylabel('Выполнено зачёт')
plt.title('Итоговая оценка')

plt.style.use('seaborn-paper')
```
